## Premier League 2020 Data 
####(Web Scrapping)

```python
import requests
```

```python
url="https://www.skysports.com/premier-league-table"
```

```python
page=requests.get(url)
```

```python
print(page.status_code)
```

```python
print(page)
```

```python
#print(page.text)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup.prettify)
```

```python
print(soup.find('a'))   #print only first a tag element
```

```python
print(soup.find_all('a'))
```

```python
soup.find_all('a')[1]
```

```python
league = soup.find('table', class_='standing-table__table')
```

```python
print(league)
```

```python
league_table = league.find_all('tbody')
```

```python
print(league_table)
```

```python
for league_teams in league_table:
    rows = league_teams.find_all('tr')
    for row in rows:
        team_names=row.find('td', class_ = 'standing-table__cell standing-table__cell--name').text.strip()
        team_points=row.find_all('td', class_ = 'standing-table__cell')[9].text.strip()
        print(team_names,' ', team_points)
        
```

```python
import pandas as pd
```

```python
league_2020 =[]

for league_teams in league_table:
    rows = league_teams.find_all('tr')
    for row in rows:
        team_names=row.find('td', class_ = 'standing-table__cell standing-table__cell--name').text.strip()
        team_points=row.find_all('td', class_ = 'standing-table__cell')[9].text.strip()
        #print(team_names,' ', team_points)
        league_dict={'Team': team_names,
                    'pts': team_points}
        league_2020.append(league_dict)
```

```python
df = pd.DataFrame(league_2020)
```
